```python
col=["person_gender","person_education","person_home_ownership","loan_intent","previous_loan_defaults_on_file"]
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in col:
    df[i]=le.fit_transform(df[i])
```

```python
df=df.astype(int)
```

```python
cols=["person_age","person_education","person_emp_exp","person_income","person_home_ownership","loan_amnt","loan_intent","loan_int_rate","loan_int_rate","loan_percent_income","cb_person_cred_hist_length","credit_score"]
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

```

```python
df
```

```python
df=df.drop(["person_gender","person_education","person_age","credit_score","loan_intent","cb_person_cred_hist_length"],axis=1)
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),annot=True)
```

```python
x=df.drop(["loan_status"],axis=1)
y=df["loan_status"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
x_train_scaler=scaler.fit_transform(x_train)
x_test_scaler=scaler.fit_transform(x_test)
```

```python
x_train_scaler
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,recall_score,precision_score
```

```python
models= {
    "logistic_regression":LogisticRegression(),
    "KNN":KNeighborsClassifier(),
    "Neive_bayes":GaussianNB(),
    "Decision_tree":DecisionTreeClassifier(),
    "SVM":SVC(probability=True)
}
result=[]
```

```python
for name,model in models.items():
    model.fit(x_train_scaler,y_train)
    y_pred= model.predict(x_test_scaler)
    acc=accuracy_score(y_test,y_pred)
    recall=recall_score(y_test,y_pred)
    pre=precision_score(y_test,y_pred)
    result.append({
        "model":name,"accuracy":round(acc,4),'recall':round(recall,4),'precision':round(pre,4)})
```

```python
result
```

```python
from sklearn.ensemble import RandomForestClassifier
random=RandomForestClassifier()
random.fit(x_train_scaler,y_train)
```

```python
y_pred_random=random.predict(x_test_scaler)
```

```python
acc=accuracy_score(y_test,y_pred_random)
r=recall_score(y_test,y_pred_random)
p=precision_score(y_test,y_pred_random)
```

```python
result.append({
        "model":"random forest","accuracy":round(acc,4),'recall':round(r,4),'precision':round(p,4)})
```

```python
result
```

```python
from xgboost import XGBClassifier
xg=XGBClassifier()
xg.fit(x_train_scaler,y_train)
```

```python
y_pred_xg=xg.predict(x_test_scaler)
```

```python
acc=accuracy_score(y_test,y_pred_xg)
r=recall_score(y_test,y_pred_xg)
p=precision_score(y_test,y_pred_xg)
```

```python
result.append({
        "model":"XG Boost","accuracy":round(acc,4),'recall':round(r,4),'precision':round(p,4)})
result
```
